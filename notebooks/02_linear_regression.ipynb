This notebook trains a **Linear Regression** baseline model on the preprocessed UrbanAirNet dataset.
Results (MAE, RMSE) will be used as the comparison baseline for Random Forest and Gradient Boosting.

## 1. Imports & Load Preprocessed Data

```python
import sys, os, json
sys.path.append('../src')
from preprocessing import load_and_preprocess

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error

os.makedirs('../report/figures', exist_ok=True)

X_train, X_test, y_train, y_test, feature_names = load_and_preprocess()
print(f'\nTrain size: {X_train.shape[0]:,}  |  Test size: {X_test.shape[0]:,}  |  Features: {X_train.shape[1]}')
```

## 2. Train Linear Regression

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
print('✅ Model trained.')
print(f'   Intercept : {lr.intercept_:.4f}')
print(f'   Number of coefficients: {len(lr.coef_)}')
```

## 3. Evaluate on Test Set

```python
y_pred = lr.predict(X_test)

mae  = mean_absolute_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)

print('=' * 45)
print('  LINEAR REGRESSION BASELINE RESULTS')
print('=' * 45)
print(f'  MAE  : {mae:.4f}')
print(f'  RMSE : {rmse:.4f}')
print('=' * 45)
```

## 4. Predicted vs Actual Scatter Plot

```python
plt.figure(figsize=(7, 6))
plt.scatter(y_test, y_pred, alpha=0.25, s=5, color='steelblue')
min_val = min(y_test.min(), y_pred.min())
max_val = max(y_test.max(), y_pred.max())
plt.plot([min_val, max_val], [min_val, max_val], 'r--', linewidth=2, label='Perfect prediction')
plt.xlabel('Actual AQI', fontsize=12)
plt.ylabel('Predicted AQI', fontsize=12)
plt.title(f'Linear Regression: Predicted vs Actual AQI\nMAE={mae:.3f}  RMSE={rmse:.3f}', fontsize=13)
plt.legend()
plt.tight_layout()
plt.savefig('../report/figures/lr_scatter.png', dpi=150)
plt.show()
print('Saved: report/figures/lr_scatter.png')
```

## 5. Time-Series Plot (First 200 Test Points)

```python
n = 200
plt.figure(figsize=(14, 4))
plt.plot(y_test[:n],  label='Actual AQI',    color='black',     linewidth=1.2)
plt.plot(y_pred[:n],  label='LR Predicted',  color='steelblue', linewidth=1.0, alpha=0.85, linestyle='--')
plt.title(f'Linear Regression — First {n} Test Predictions', fontsize=13)
plt.xlabel('Hour Index')
plt.ylabel('AQI')
plt.legend()
plt.tight_layout()
plt.savefig('../report/figures/lr_timeseries.png', dpi=150)
plt.show()
print('Saved: report/figures/lr_timeseries.png')
```

## 6. Feature Coefficients (Top 15 by Magnitude)

```python
coef = lr.coef_
sorted_idx = np.argsort(np.abs(coef))[::-1][:15]

plt.figure(figsize=(10, 5))
colors = ['steelblue' if c > 0 else 'coral' for c in coef[sorted_idx]]
plt.bar(range(len(sorted_idx)), coef[sorted_idx], color=colors, edgecolor='black')
plt.xticks(range(len(sorted_idx)),
           [feature_names[i] for i in sorted_idx],
           rotation=45, ha='right')
plt.title('Linear Regression — Top 15 Coefficients (by magnitude)', fontsize=13)
plt.ylabel('Coefficient Value')
plt.axhline(0, color='black', linewidth=0.8)
plt.tight_layout()
plt.savefig('../report/figures/lr_coefficients.png', dpi=150)
plt.show()
print('Saved: report/figures/lr_coefficients.png')
```

## 7. Save Results for Comparison Notebook

```python
lr_results = {
    'model': 'Linear Regression',
    'mae':  round(float(mae),  4),
    'rmse': round(float(rmse), 4),
    'note': 'Baseline — no hyperparameter tuning'
}

with open('../report/lr_results.json', 'w') as f:
    json.dump(lr_results, f, indent=2)

print('\n✅ Results saved to report/lr_results.json')
print(lr_results)
print('\n→ Share this number with John and Connor as the baseline to beat.')
```
