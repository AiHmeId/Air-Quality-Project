Trains a **Linear Regression** baseline to predict `Health_Risk_Score`.  
Results (MAE, RMSE) serve as the comparison baseline for Random Forest and Gradient Boosting.

## 1. Imports & Load Preprocessed Data

```python
import sys, os, json
sys.path.append('../src')
from preprocessing import load_and_preprocess

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error

os.makedirs('../report/figures', exist_ok=True)
os.makedirs('../report', exist_ok=True)

X_train, X_test, y_train, y_test, feature_names = load_and_preprocess()
print(f'\nTrain: {X_train.shape[0]:,} rows  |  Test: {X_test.shape[0]:,} rows  |  Features: {X_train.shape[1]}')
```

## 2. Train Linear Regression

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
print('✅ Model trained.')
print(f'   Intercept            : {lr.intercept_:.4f}')
print(f'   Number of features   : {len(lr.coef_)}')
```

## 3. Evaluate on Test Set

```python
y_pred = lr.predict(X_test)

mae  = mean_absolute_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)

print('=' * 50)
print('  LINEAR REGRESSION BASELINE — RESULTS')
print('=' * 50)
print(f'  MAE  : {mae:.4f}')
print(f'  RMSE : {rmse:.4f}')
print('=' * 50)
print()
print('Health_Risk_Score range in test set:', f'{y_test.min():.2f} – {y_test.max():.2f}')
print(f'MAE as % of range: {mae / (y_test.max() - y_test.min()) * 100:.1f}%')
```

## 4. Predicted vs Actual Scatter Plot

```python
plt.figure(figsize=(7, 6))
plt.scatter(y_test, y_pred, alpha=0.5, s=20, color='steelblue', edgecolors='white', linewidths=0.3)
min_val = min(y_test.min(), y_pred.min()) - 0.1
max_val = max(y_test.max(), y_pred.max()) + 0.1
plt.plot([min_val, max_val], [min_val, max_val], 'r--', linewidth=2, label='Perfect prediction')
plt.xlabel('Actual Health Risk Score', fontsize=12)
plt.ylabel('Predicted Health Risk Score', fontsize=12)
plt.title(f'Linear Regression: Predicted vs Actual\nMAE={mae:.3f}   RMSE={rmse:.3f}', fontsize=13)
plt.legend()
plt.tight_layout()
plt.savefig('../report/figures/lr_scatter.png', dpi=150)
plt.show()
print('Saved → report/figures/lr_scatter.png')
```

## 5. Time-Series Plot (All 200 Test Points)

```python
plt.figure(figsize=(14, 4))
plt.plot(y_test,  label='Actual Health Risk Score', color='black',     linewidth=1.5)
plt.plot(y_pred,  label='LR Predicted',             color='steelblue', linewidth=1.2, alpha=0.85, linestyle='--')
plt.title('Linear Regression — Test Set Predictions (200 records)', fontsize=13)
plt.xlabel('Test Record Index')
plt.ylabel('Health Risk Score')
plt.legend()
plt.tight_layout()
plt.savefig('../report/figures/lr_timeseries.png', dpi=150)
plt.show()
print('Saved → report/figures/lr_timeseries.png')
```

## 6. Top Feature Coefficients

```python
coef = lr.coef_
sorted_idx = np.argsort(np.abs(coef))[::-1][:15]

bar_colors = ['steelblue' if c > 0 else 'coral' for c in coef[sorted_idx]]

plt.figure(figsize=(11, 5))
plt.bar(range(len(sorted_idx)), coef[sorted_idx], color=bar_colors, edgecolor='black')
plt.xticks(range(len(sorted_idx)),
           [feature_names[i] for i in sorted_idx],
           rotation=45, ha='right', fontsize=9)
plt.axhline(0, color='black', linewidth=0.8)
plt.title('Linear Regression — Top 15 Feature Coefficients\n(blue = positive, red = negative)', fontsize=12)
plt.ylabel('Coefficient Value')
plt.tight_layout()
plt.savefig('../report/figures/lr_coefficients.png', dpi=150)
plt.show()
print('Saved → report/figures/lr_coefficients.png')

print('\nTop 10 features by coefficient magnitude:')
for i in sorted_idx[:10]:
    print(f'  {feature_names[i]:<25}  coef = {coef[i]:+.4f}')
```

## 7. Save Results for Comparison Notebook

```python
lr_results = {
    'model': 'Linear Regression',
    'mae':   round(float(mae),  4),
    'rmse':  round(float(rmse), 4),
    'note':  'Baseline — no hyperparameter tuning'
}

with open('../report/lr_results.json', 'w') as f:
    json.dump(lr_results, f, indent=2)

print('✅ Saved → report/lr_results.json')
print(lr_results)
print()
print('→ Share MAE and RMSE with John and Connor as the baseline to beat!')
```
